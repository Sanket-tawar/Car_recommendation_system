```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
base_url = 'https://www.cars24.com/buy-used-cars-pune/'
cars_list = []
page_num = 1
max_pages = 50
```

```python
while page_num <= max_pages:
    if page_num == 1:
        url = base_url
    else:
        url = f"{base_url}?page={page_num}"

    print('Scraping page no. :',page_num)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    car_div = soup.find_all('a', class_='styles_carCardWrapper__sXLIp')
    if not car_div:
        break
    for car in car_div:
        car_data = {}
        car_data['Car Url'] = car.get('href', None)
        car_div_temp = car.find('div', class_='styles_normalCardWrapper__qDZjq')
        if not car_div_temp:
            continue
        img_tag = car_div_temp.find('img', class_='shrinkOnTouch')
        car_data['Image Url'] = img_tag.get('src') if img_tag else None
        car_name = car_div_temp.select_one('span.jtaHBC')
        car_var = car_div_temp.select_one('span.kTVtGZ')
        car_price = car_div_temp.select_one('p.bdSJhg')
        car_loc = car_div_temp.select_one('div.styles_ellipsis__uatjG p.kTVtGZ')
        car_assured = car_div_temp.select_one('span.ecaxpZ')
        car_data['Name'] = car_name.get_text(strip=True) if car_name else None
        car_data['Variant'] = car_var.get_text(strip=True) if car_var else None
        car_data['Price'] = car_price.get_text(strip=True) if car_price else None
        car_data['Location'] = car_loc.get_text(strip=True) if car_loc else None
        car_data['Assured'] = car_assured.get_text(strip=True) if car_assured else None

        emi_text = None
        pricing_div = car_div_temp.select_one('div.styles_pricingDetail__Q_3hz')
        if pricing_div:
            for tag in pricing_div.find_all(['p','span']):
                if 'EMI' in tag.get_text(strip = True):
                    emi_text = tag.get_text(strip = True)
                    break
        car_data['EMI'] = emi_text
        
        details = [p.get_text(strip = True) for p in car_div_temp.select('p.eTYSzW')]
        car_data['KM driven'] = details[0] if len(details) > 0 else None
        car_data['Fuel Type'] = details[1] if len(details) > 1 else None
        car_data['Transmission'] = details[2] if len(details) > 2 else None
        car_data['Registration'] = details[3] if len(details) > 3 else None  
        
        cars_list.append(car_data)

    page_num += 1
    time.sleep(2)
```

```python
df = pd.DataFrame(cars_list)
df
```

```python
df.to_csv('Cars24 Data.csv',index = False)
```
