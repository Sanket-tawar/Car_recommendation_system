```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Admin\\Car_recommendation_system\\Cars24 Data.csv")
df.head(50)
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df['price'] = df['price'].astype(str).str.replace('₹','', regex = False).str.replace('lakh','', regex = False).astype(float) * 100000
df['km driven'] = df['km driven'].astype(str).str.replace('k','', regex = False).str.replace('L','', regex = False).str.replace('km','', regex = False).str.replace('m','',regex=False).astype(float)
df['emi'] = df['emi'].astype(str).str.replace('EMI ₹', '', regex = False).str.replace('/m*', '',regex = False).str.replace(',','', regex = False).astype(float)
```

```python
df.to_csv("C:\\Users\\Admin\\Car_recommendation_system\\processed_data(cars24).csv",index = False)
```
